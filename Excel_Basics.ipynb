Q1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.

You have the following dataset:
image.png

 Answer - 1) Relative Cell Reference (B2)

> Changes automatically when copied.

Example: In C2 write =B2*10%
Copy down → becomes =B3*10%, =B4*10%

Use when the formula should adjust row/column-wise.

Logic: “Move with me.”

- 2) Absolute Cell Reference ($B$2)

> Never changes when copied.

Example: Suppose B6 = 10% tax rate
In C2 write =B2*$B$6
Copy down → always multiplies by B6.

Use for constants (tax, rate, fixed value).

Logic: “Stay fixed, no matter what.”

- 3) Mixed Cell Reference ($B2 or B$2)

> Partially fixed.

Example: =$B2*10%

Column B fixed

Row changes when copied down

Useful in tables, comparisons, or matrix-style formulas.

Logic: “Fix one direction, allow the other.”
